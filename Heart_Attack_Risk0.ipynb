```python
df['Gender']=df['Gender'].replace({'Male':0,'Female':1})
```

```python
def PAL(x):
  if x=='Moderate':
    return 0
  elif x=='Low':
    return 1
  elif x=='High':
    return 2
df['Physical_Activity_Level']=df['Physical_Activity_Level'].apply(PAL)
```

```python
df.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
labelfit=le.fit(df['Stress_Level'])
df['Stress_Level']=labelfit.transform(df['Stress_Level'])
```

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
labelfit1=le1.fit(df['Heart_Attack_Risk'])
df['Heart_Attack_Risk']=labelfit1.transform(df['Heart_Attack_Risk'])
```
